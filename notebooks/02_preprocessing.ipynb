```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../data/raw/Indian_Kids_Screen_Time.csv")
df.head()
```

Handle Missing Values

```python
df["Health_Impacts"] = df["Health_Impacts"].fillna("None")
```

```python
df.isnull().sum()
```

```python
df["Gender"] = df["Gender"].str.strip()
df["Primary_Device"] = df["Primary_Device"].str.strip()
df["Urban_or_Rural"] = df["Urban_or_Rural"].str.strip()
```

Age Bands.

```python
df["Age_Band"] = pd.cut(
    df["Age"],
    bins=[0,5,9,13,17],
    labels=["3-5","6-9","10-13","14-17"]
)
```

```python
df[["Age","Age_Band"]].head()
```

Screen Time Levels

```python
df["Screen_Time_Level"] = pd.cut(
    df["Avg_Daily_Screen_Time_hr"],
    bins=[0,2,4,6,24],
    labels=["Low","Moderate","High","Very High"]
)
```

```python
df[["Avg_Daily_Screen_Time_hr","Screen_Time_Level"]].head()
```

Save Clean Dataset for dashboard preparation

```python
df.to_csv("../data/processed/screensense_clean.csv", index=False)
```
